# 2. Replacing Missing Values

The purpose is to replace the missing values of the dataset by alternative ones. 
Given that our dataset does not include missing values, we've decided to delete cells and create our owns. 

```python
# 0. Import librairies & Dataset

import pandas
import numpy as np
df = pandas.read_csv("overbuff.csv", encoding = "latin1")
df.isnull().sum()

# 1. Replace some values by a null ones #

df = df.replace(2.75, np.nan)
df.isnull().sum()
```

```python
# 2. Replace the missing values by the mean of each class of Hero (Tank, support...)

df_Role = df.Role.unique()


for i in df_Role : 
    print("\n")
    print(i)
    
    cls = df.loc[(df['Role'] == i)]
    print("*****************************")
    
    
    avrg = cls.loc[:,"Pick_rate"].mean()
    print("Pick_rate")
    print(avrg)
      
    df.loc[(df['Role'] == i) & (df["Pick_rate"].isnull()), "Pick_rate"]=avrg
    
    avrg = cls.loc[:,"Tie_Rate"].mean()
    print("Tie_rate")
    print(avrg)
    
    df.loc[(df['Role'] == i) & (df["Tie_Rate"].isnull()), "Tie_Rate"]=avrg
    
    
    avrg = cls.loc[:,"On_fire"].mean()
    print("On_fire")
    print(avrg)
    
    df.loc[(df['Role'] == i) & (df["On_fire"].isnull()), "On_fire"]=avrg
```

```python
# 3. Check is there is any remaining missng values #

df.isnull().sum()
```

## To conclude the n°2

In this part, we've created some missing values in our dataset and then replace them by the mean of their columns in orer to obtain a relevant dataset. This technique can be applies in any table containing missing values.
